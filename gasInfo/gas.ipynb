```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.title('Gráfico de aumento de preços em US$')

list_country = ['Italy', 'UK', 'USA', 'Mexico']
for country in gas: 
    if country in list_country:
        plt.plot(gas['Year'], gas[country], label = [country], marker='.')
        
  #Adição dos preços por país separados por cores específicas      
# plt.plot(gas['Year'], gas['Italy'], 'y.-', label = 'Itália')
# plt.plot(gas['Year'], gas['UK'], 'r.-', label = 'Reino Unido')
# plt.plot(gas['Year'], gas['Germany'], 'c.-', label = 'Germânia')
# plt.plot(gas['Year'], gas['USA'], 'b.-', label = 'EUA')
# plt.plot(gas['Year'], gas['Mexico'],'g.-', label = 'México')

plt.xticks(gas['Year'][::2])
plt.xlabel('Anos')
plt.ylabel('US$')

plt.legend()
plt.savefig('gasInfo.pdf')
plt.show()

```
